## Clasical Information
- Consider a physical system that stores information: let us call it X
- Assume X can be in any one of a number of finite classical states at each moment
    - Can describe X with absolute certainty and NO ERROR
    - Denoted Sigma(Σ)
    - E.x (X is a bit, its classical state set is Σ = {0,1} ) // (X is a die, Σ={1,2,3,4,5,6} )

- In the case of uncertainty in a classical system, we can have a probability of a certain state. We can refer to this as **Probabalistic State**
    - E.x (X is a bit with a 3/4 prob. of being a 0 and 1/4 prob. of being a 1)
        - Pr(X=0) = 3/4 and Pr(X=1)=1/4
    - Can also be represented as COLUMN VECTOR, where prob. corresponds to the number written in each component of the column vector (3/4  1/4) <- **Probability vector**

## Dirac Notation  ⟨|⟩
- classical set X has |Σ| states in it
    - The states have to be ordered
- Any Choice from the set of states can be represented by |*a*⟩. Ket
    - This representation corresponds to a column vector with a 1 in the a'th entry and a 0 for all other entries
        - e.x in the classical bit example, Σ = {0,1}. |0⟩ means the 0 compoenent is 1, (1,0).
        - for the dice example, Σ={1,2,3,4,5,6}. |5⟩ means the 5 compoenent is true, (0,0,0,0,1,0) <- Standard Basis Vectors
        - If we have probabilistic state, multiplying the Ket by the probability is how you represent it

#### Part 2
- ⟨a| is a row vector where there is a 1 for the state in the row and a 0 for all the other states
    - "bra a"

- Multiplying a row vector by a column vector will give us 1 number. think of it as a scalar
- ⟨a| * |*b*⟩ can either equal 0 or 1 ---- ⟨a|b⟩ 
    - if its a zero, then the classical states are different and a 1 if they are the same

- Multiplying a column vector by a row vector gives us a matrix
- |a⟩⟨b| will result in a matrix where the (a,b) entry is a 1 and all others are 0

## Deterministic Operations
- modeled by functions f
- f Σ -> Σ. Transforms the classical state *a* into f(*a*) for each of a in Σ
- any of these deterministic operation functions creates a new matrix M that satisfies
    - M | *a*> = f(a)> for all a, or all states
- M(b,a) is a matrix with a 1 for where b = f(a) and a zero for b != f(a)
    - "Row b, Column a)

- If we have a system in a probabalistc state, and we perform a deterministic operation on that system, then the resulting **Probabalistic** state can be obtained by matrix-vector multiplication
    - v |-> M*v
    - |-> represents the mapping of that vector into the product

## Quantum Information
- A quantum state of a system is represented by a column vector whose indices are placed in corrospondence with the classical states of that system
    - Entries in the system are complex numbers rather than non-negative real numbers
    - the sum of the absolute values squared of the entries must equal 1
- The **Euclidean Norm** of a vector with complex numbers is the square root of the sum of the absolute values squared of the entries of the vecor
    - sum =0
    for entries in vector:
        sum += |entries|^2
    sum = sqrt(sum)
- Quantum states are unit vectors with respect to the l2 norm.
-written as ||v||

- Examples of quantum state
    - a qubit has standard basis states |0⟩ and |1⟩
    - the entries to these in the column vector representation are complex numbers (in the basis states the complex components are 0)
    - there are also the plus state and the minus state. |-⟩ and |+⟩
        - the entries are complex numbers (as with all qubits)
        - |-⟩ = $1/ \sqrt{2}$ |0⟩ - $1/ \sqrt{2}$ |1⟩
        - |+⟩ = $1/ \sqrt{2}$ |0⟩ + $1/ \sqrt{2}$ |1⟩
    - random state $(1+2i)/3$ |0⟩ - 2/3 |1⟩
        - the noem of first coefficient is 5/9 and the second is 4/9, summing to 1

## Unitare Operations
- Operations on quatum state vecors are called unitary matrices. 
- Multiplying a matrix by its conjugate transpose on either side to get the identity matrix

-Unitary matrices are the only matrices that transforms quantum state vectors to other quantum state vectors (just like ** the other)


Any matrix that equals its own conjugate teanspose means that its a hermitian matrix

A special type of hermitian matrix is the hadamard operation denoted 
  $$\begin{bmatrix} 1 sqrt 2 & 1 sqrt 2 \\ 1 sqrt 2 &- 1 sqrt 2 \end{bmatrix}$$

  the hadamard op. applied to ket 0 or ket 1 will give the plus and minus states, and vice versa, respectively


A phase operation is another special matrix wich takes the form
  $$\begin{bmatrix} 1 & 0 \\ 0  & e^i0 \end{bmatrix}$$

can take many famous forms of different operations (s gate, t gate)

Unitary operations dont need a matrix representation to calculate their results. they can be kept in their original representations and computed algebraically, assuming we know what the result will be applied to a standard basis
