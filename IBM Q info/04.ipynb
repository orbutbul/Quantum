## Quantum Teleportation

Q. Teleportaion is the transmission of information through Q. entanglement.

Scenario:
- Alice has a qubit Q that she wants to transmit to Bob
- Alice cant send the qubit, she can send classical information to Bob
- Alice and Bob share an e-bit
- The state of Q is unknown to Alice and Bob
- Q might be entangled with other systems and needs to stay that way
- if Bob recieves the transmission, Alice does not have the Qubit in its original state

- Is there any way to transfer a qubit(quantum information) without the e-bit or without quantum information (using only classical information)?
    - No lol
    - No cloning theorem means not all qunatum states can be copied

## Superdense Coding

- Allows for 2 classical bits of communication to be packed into one qubit, at the cost of one e-bit

Alice has 2 classical bits that she wants to send to bob,
alice has sent 1 qubit to bob, so they share an e-bit

without an e-bit, this task would be impossible (holevos theorem - you cant send 2 classical bits of info. with 1 qubit)

![title](15.png)

if a,b are zero, nothing gets flipped (represented by the identity matrix), the state stays the same 
if a=0, b=1, the phase gets flipped (identity matrix tensored with the Z gaete), the phi plus state becomes a phi minus state
if a=1,b=0, the state is swapped, phi plus becomes psi plus
if a=1,b=1, the state and sign swap, phi plis becomes psi minus

#### Remarks
- superdense coding is not useful in a practical sense
- the underlying idea is important and gives more info on entanglement
- together with teleportation, superdense coding creates an equivalence
    - at the expense of 1 e-bit, we can communicate 1 qubit of quantum information or 2 bits of classical information at the cost of the other

## CHSH Game

Not an actual game but an abstraction to illustrate an idea

the CHSH game is an example of a **Non-Local** game
- Alice and Bob work together, they either win together or lose together
- Alice and Bob can communicate before the game starts, but once it starts thet can not communicate

- Alice and bob are each asked 1 question( Question X for Alice and Y for Bob).
- Their answers (a and b) are then given to the referee
    - in the case of the CHSH game, the questions are bits and the answers are bits
- The referee decides if they win by if their answers matched the questions X and Y in a particular way

- The referee use randomnes to ask X and Y, so they dont know what they or the other will be asked (prisoner dilemna energy)

- in summary, Alice and bob win if their answers (a,b) are in agreement with the questions (x,y) according to some fixed rule
    - it is assumed that Alice and Bob Know the rule
